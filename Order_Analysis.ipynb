# I. Overall Business Metrics


```python
 # 1. Total Revenue (Total Order Value)

Total_Revenue=df['Order Value'].sum()
print('Total_Revenue: ', Total_Revenue)
```

```python
# 2. Total Net Profit

Total_Net_Profit = df['Profit'].sum()
print('Total_Net_Profit:',Total_Net_Profit)
```

```python
# 3. Profit Margin Percentage

Profit_Margin_Percentage = (Total_Net_Profit/Total_Revenue)*100
print('Profit_Margin_Percentage: ', Profit_Margin_Percentage)
```

```python
# 4. Average Order Value

Average_Order_Value = df['Order Value'].mean()
print('Average_Order_Value:', Average_Order_Value)
```

# II. Performance Breakdown

```python
#● By Geographic Region

Region_Summary =df.groupby('Region')[['Order Value', 'Profit']].sum().reset_index()
Region_Summary
```

```python
Region_Summary.plot(x= 'Region', y=['Order Value', 'Profit'], kind='bar')
plt.title('Region_Summary')
plt.xlabel('Region')
plt.ylabel('Amount')
plt.show()
```

```python
# Summary by Product

Product_Revenue = df.groupby('Product_Name')[['Order Value', 'Profit']].sum().reset_index()
Product_Revenue
```

```python
Product_Revenue.plot(x='Product_Name', y=['Order Value', 'Profit'], kind='bar', figsize=(20,5))

plt.title('Product based Revenue')
plt.xlabel('Products')
plt.ylabel('Amount')
plt.show()
```

```python
# By Product Category

Product_Categories= df.groupby('Product_Category')[['Order Value', 'Profit']].sum().reset_index()
Product_Categories
```

```python
Product_Categories.plot(x='Product_Category', y=['Order Value', 'Profit'], kind='bar', figsize=(10,5))

plt.title('Category based Revenue')
plt.xlabel('category name')
plt.ylabel('Amount')
plt.show()
```

```python
# By Customer Segment

Segment_Wise_revenue= df.groupby('Segment')[['Order Value', 'Profit']].sum().reset_index()
Segment_Wise_revenue
```

```python
Segment_Wise_revenue.plot(x='Segment', y=['Order Value', 'Profit'], kind='bar', figsize=(10,5))

plt.title('Segment based Revenue')
plt.xlabel('Segment_Name')
plt.ylabel('Amount')
plt.show()
```

# III. Top 10 Rank Products

```python
# By Products

Product_wise_Values= df.groupby('Product_Name')[['Order Value', 'Profit']].sum(). sort_values(by='Order Value',ascending=False).reset_index()
Product_wise_Values.head(10)
```

# IV Time-Series Trend - Monthly Performance Trend Analysis


```python
df['Order Date']=pd.to_datetime(df["Order Date"])

df["Month"] = df["Order Date"].dt.month_name()

Monthly_Performance= df.groupby('Month')[['Order Value', 'Profit']].sum().sort_values('Order Value', ascending=False).reset_index()
Monthly_Performance
```

```python
# Monthly trend

Monthly_Performance.plot(x='Month', y=['Order Value', 'Profit'], kind='line', marker='o')
plt.title("Monthly Revenue and Profit Trend")
plt.xlabel("Month")
plt.ylabel("Amount")
plt.show()
```
